```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apps = pd.read_csv("apps.csv")
reviews = pd.read_csv("user_reviews.csv")

apps.head()
```

```python
reviews.head()
```

```python
apps.dropna(inplace=True)

apps['Installs'] = apps['Installs'].str.replace('+','')
apps['Installs'] = apps['Installs'].str.replace(',','')
apps['Installs'] = apps['Installs'].astype(int)

apps['Reviews'] = pd.to_numeric(apps['Reviews'])
apps['Rating'] = pd.to_numeric(apps['Rating'])
```

```python
apps['Category'].value_counts().head(10).plot(kind='bar')
plt.title("Top Categories")
plt.show()
```

```python
sns.histplot(apps['Rating'], bins=20)
plt.title("Ratings Distribution")
plt.show()
```

```python
apps['Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Free vs Paid Apps")
plt.show()
```

```python
reviews.dropna(inplace=True)

reviews['Sentiment'].value_counts().plot(kind='bar')
plt.title("User Sentiments")
plt.show()
```

```python
merged_data = pd.merge(apps, reviews, on="App")

merged_data.head()
```

Recommendation

1. Developers should focus on high-demand categories such as Games, Tools, and Productivity to reach larger audiences.
2. Apps with better ratings and reviews attract more installs, so improving user experience, performance, and design is essential.
3. Since most users prefer free apps, developers can use in-app purchases or ads instead of paid downloads.
4. Regular updates and addressing user complaints can significantly improve app ratings and satisfaction.
5. Monitoring user sentiment helps identify problems early and enhance the overall app quality.

Conclusion

The analysis of Google Play Store data shows that most apps are free, and categories like Games, Tools, and Productivity are the most popular. Apps with higher ratings and positive user sentiment receive more downloads, indicating that user experience and quality strongly influence app success. By studying both app data and user reviews, developers can better understand market trends and improve their apps to match user expectations.
