```python
from textblob import TextBlob
import re

# Dataset
dataset = [
"hello","help","helmet","healthy","heart","heat",
"computer","company","complete","comparison",
"data","database","dataset","data science",
"machine","machine learning","model","method",
"program","project","prediction","process"
]

# Preprocessing
clean_data = []
for word in dataset:
    word = word.lower()
    word = re.sub(r'[^a-z ]','', word)
    clean_data.append(word)

# Autocomplete function
def autocomplete(prefix):
    suggestions = []
    for word in clean_data:
        if word.startswith(prefix):
            suggestions.append(word)
    return suggestions

# Input
user_input = input("Enter word: ")  

# Autocomplete
print("Autocomplete Suggestions:", autocomplete(user_input))

# Autocorrect
blob = TextBlob(user_input)
print("Autocorrect Word:", blob.correct())
```

```python
import pandas as pd
import re

dataset = [
"hello","help","helmet","healthy","heart","heat",
"computer","company","complete","comparison",
"data","database","dataset","data science",
"machine","machine learning","model","method",
"program","project","prediction","process"
]

df = pd.DataFrame(dataset, columns=["words"])
df.head()
```

```python
df["clean_words"] = df["words"].str.lower()
df["clean_words"] = df["clean_words"].apply(lambda x: re.sub(r'[^a-z ]','',x))
df.head()
```

```python
df["length"] = df["clean_words"].apply(len)
df.head()
```

```python
import matplotlib.pyplot as plt

df["length"].value_counts().plot(kind="bar")
plt.title("Word Length Distribution")
plt.xlabel("Word Length")
plt.ylabel("Count")
plt.show()
```
